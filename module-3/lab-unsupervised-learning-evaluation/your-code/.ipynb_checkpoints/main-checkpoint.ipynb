```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


kmeans = KMeans(n_clusters=8).fit(X)

score = silhouette_score(X, kmeans.labels_, metric='euclidean')

score
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


kmeans = KMeans(n_clusters=5).fit(X)

score = silhouette_score(X, kmeans.labels_, metric='euclidean')

score
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


kmeans = KMeans(n_clusters=3).fit(X)

score = silhouette_score(X, kmeans.labels_, metric='euclidean')

score
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from yellowbrick.cluster import KElbowVisualizer

visualizer = KElbowVisualizer(
    KMeans().fit(X),
    k = (3, 11), metric='silhouette', #range of values for k (min is 2)
)

visualizer.fit(X)
visualizer.poof()
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
from sklearn.cluster import AgglomerativeClustering

visualizer = KElbowVisualizer(
    AgglomerativeClustering().fit(X),
    k = (3, 11), metric='silhouette', #range of values for k (min is 2)
)

visualizer.fit(X)
visualizer.poof()
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardizing the features
x = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)


# pca = PCA(n_components=2)
# x = pca.fit(X)

# sns.scatterplot(
#         data = x,
#     )

# plt.show()
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
# principalDf.head()
finalDf = pd.concat([principalDf, y], axis = 1)

finalDf.head()
```

```python
fig = plt.figure(figsize = (8,8))

ax = fig.add_subplot(1,1,1) 

colors = ['r', 'g', 'b']
for target, color in zip(y,colors):
    indicesToKeep = finalDf[0] == y
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(y)
ax.grid()
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 4, figsize = (15, 7))

for i, p in enumerate([5, 30, 50, 100]):
    tsne = TSNE(perplexity=p)
    reduced = pd.DataFrame(tsne.fit_transform(
        X
    ))
    sns.scatterplot(
        data = reduced,
#         x = 0,
#         y = 1,
#         hue = df['species_id'],
        ax = ax[i]
    )
    ax[i].set_title(f"Perplexity = {p}")
    
plt.show()
```
